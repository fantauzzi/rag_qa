```python
starting_year = 2022
```

```python
metadata = pd.DataFrame()

metadata['title'] = title_basics.primaryTitle
metadata['year'] = pd.to_numeric(title_basics.startYear, errors='coerce')
metadata['title_type'] = title_basics.titleType
metadata['id'] = title_basics.tconst
metadata = metadata[metadata.year.notna() & (metadata.year >= starting_year) & (metadata.title_type == 'movie')]
metadata.year = metadata.year.astype(int)
metadata
```
